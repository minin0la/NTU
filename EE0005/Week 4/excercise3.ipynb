```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
data = pd.read_csv('train.csv')
```

```python
houseCatData = data[['MSSubClass', 'Neighborhood', 'BldgType', 'OverallQual']]
```

```python
houseCatData.head()
```

```python
houseCatData.head()
```

```python
for x in houseCatData:
    houseCatData[x] = houseCatData[x].astype('category')
houseCatData.info()
```

```python
saleprice = data[['SalePrice']]
housecatprice = pd.concat([houseCatData, saleprice], axis=1)
housecatprice.head()
```

```python
f, axes = plt.subplots(1,2, figsize=(40,20))

sb.boxplot(x='SalePrice', y='BldgType', data=housecatprice, ax=axes[0])
sb.boxplot(x='SalePrice', y='Neighborhood', data=housecatprice, ax=axes[1])
f, axes = plt.subplots(1,1, figsize=(40,20))
sb.boxplot(x='SalePrice', y='MSSubClass', data=housecatprice)
```
